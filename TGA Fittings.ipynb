Kinetics of Thermal Decomposition of Model Systems
==

1.  Global Package Imports

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
from scipy import integrate
import scipy.optimize
import matplotlib.pyplot as plt
import scipy.signal
from lmfit import minimize, Parameters
```

2.  Difining Miscellaneous Functions
--

```python
def integrate_pars(p):
    extra_args = tuple(p)
    return integrate.odeint(SB_Model, initialalpha, T_K, extra_args)[:,0]

def errfun(p):
    return (Experimental_Alpha - integrate_pars(tuple(p)))

def sumofsquares(p):
    return (errfun(p)**2).sum()

def comparefit(p):
    model_alpha = integrate_pars(p)
    model_dalpha = numpy.vectorize(SB_Model)(model_alpha, T_C, *p)
    derExperimental_Alpha = scipy.signal.savgol_filter(Experimental_Alpha, window_length=55, polyorder=2, deriv=1, mode='interp')/1
    
    f, ax = plt.subplots(3, sharex=True, sharey=False)
    
    ax[0].plot(T_C, Experimental_Alpha, color='green')
    ax[0].plot(T_C, model_alpha)
    ax[0].set_ylabel(r'$\alpha$')
    
    ax[1].plot(T_C, model_alpha - Experimental_Alpha)
    ax[1].set_ylabel('Model - Data')
    ax[1].axhline(0, color='black')
    
    ax[2].plot(T_C, derExperimental_Alpha, color='green')
    ax[2].plot(T_C, numpy.gradient(model_alpha))
    ax[2].set_ylabel(r'd$\alpha/$d$t$')
    ax[2].set_xlabel('Temperature / K')
    
    ax[2].legend(['Data', 'Model'])
    
    f.subplots_adjust(hspace=0)
    
    plt.show()
    
def print_values(ps):
    for name, p in zip(names, ps):
        print name, '=', p

def residual(params):
    return errfun((params[p].value for p in names))
```

3.  Defining Functions for Synthetic Data
--

3.1.  Ordinary Gompertz Function for Sigmoidal Data


```python
def gompertz_ord(x,End_T,Lift_T,Lift_Alpha):
        c = 10.0/End_T
        b = exp(log(-log(Lift_Alpha))+c*Lift_T)
        return (exp(-b*exp(-c*x)))
```

```python
End_T = 1000
Lift_T = 200
Lift_Alpha = 0.2

x = np.linspace(1,End_T,End_T)

g = gompertz_ord(x,End_T,Lift_T,Lift_Alpha)

plt.plot(x,g)
```

3.2.  Noise Gompertz Function for Sigmoidal Data

```python
def gompertz_noise(x,End_T,Lift_T,Lift_Alpha,Noise):
        c = 10.0/End_T
        b = exp(log(-log(Lift_Alpha))+c*Lift_T)
        return (((exp(-b*exp(-c*x)))) + (Noise/100.0*np.random.ranf(len(x))))
```

```python
End_T = 1000
Lift_T = 200
Lift_Alpha = 0.2
Noise = 1

x = np.linspace(1,End_T,End_T)

g = gompertz_noise(x,End_T,Lift_T,Lift_Alpha, Noise)

plt.plot(x,g)
```

4.  Defining Kinetic Model
--

4.1  Sesstak-Bergren Model

```python
def SB_Model(alpha, T, A, E, m, n, p):
    R = 8.314
    beta = 15
    # clamp alpha
    alpha = np.max(np.min(alpha, 1 - np.spacing(1)), np.spacing(1))
    dadt = A/beta*np.exp(-E/(R*T))*alpha**m*(1 - alpha)**n*(-np.log(1 - alpha))**p
    if np.isreal(dadt) and np.isfinite(dadt):
        # no negative derivatives
        return np.max(dadt, 0)
    else:
        return 0
```

5.1  Unconstrained Optimisation on alpha using Least Squares

```python
names = ['A', 'E', 'm', 'n', 'p']
p0 = (300., 40000., 0.75, 6.0, 0)

End_T = 1000
Start_T = 25
dT = 1

Steps = (End_T-Start_T)/dT
Lift_T = 500
Lift_Alpha = 0.37

T_C = array(np.linspace(1,End_T,Steps))
T_K = T_C + 273.15

initialalpha = 0.00000005

Experimental_Alpha = array(gompertz_noise(T_C,End_T,Lift_T,Lift_Alpha, 0))
```

```python
#%%timeit
a = integrate_pars(p0)

plt.subplot(2, 1, 1)
plt.plot(T_C,a)
plt.plot(T_C,Experimental_Alpha)
```

```python
%%time
fitpars = {}
newstart = numpy.array(p0)
fitpars['Unconstrained'], cov_p, infodict, mesg, ierr = scipy.optimize.leastsq(errfun, newstart, full_output=True)

print sumofsquares(fitpars['Unconstrained'])
print "---"
print_values(p0)
print "---"
print_values(fitpars['Unconstrained'])
print "---"
comparefit(fitpars['Unconstrained'])
```

5.2 Constrained Optimisation on alpha

```python
%%time

names = ['A', 'E', 'm', 'n', 'p']

End_T = 1000
Start_T = 25
dT = 1

Steps = (End_T-Start_T)/dT
Lift_T = 500
Lift_Alpha = 0.37

T_C = array(np.linspace(Start_T,End_T,Steps))
T_K = T_C + 273.15

initialalpha = 0.1e-9

Experimental_Alpha = array(gompertz_noise(T_C,End_T,Lift_T,Lift_Alpha, 0))

params = Parameters()
params.add('A', value=250.0, min=0.001, max=2500.)
params.add('E', value=20000., min=15e0, max=50e4)
params.add('m', value=0.75, min=-10.0, max=10.0)
params.add('n', value=6.0, min=-10.0, max=10.0)
params.add('p', value=0.0, min=-10.0, max=10.0, vary=False)


out = minimize(residual, params, method='leastsq')

fitpars['Constrained'] = tuple(out.params[n].value for n in names)
print sumofsquares(fitpars['Constrained'])
print "---"
print_values(fitpars['Constrained'])
print "---"
comparefit(fitpars['Constrained'])
```

5.3  Brute Force Search on alpha

```python
%%time

names = ['A', 'E', 'm', 'n', 'p']

ranges = ((0.0001, 25000.), # A
          (15.0e1, 50.0e4), # E
          (-7.0, 7.0), # m
          (-7., 7.0), # n
          (-10.0, 10.0), # p
          )

End_T = 1000
Lift_T = 500
Lift_Alpha = 0.37

T_C = array(np.linspace(1,End_T,End_T))
T_K = T_C + 273.15

initialalpha = 0.00000005

Experimental_Alpha = array(gompertz_noise(T_C,End_T,Lift_T,Lift_Alpha, 0.0))

fitpars['Brute'], fval, grid_output, Jout = scipy.optimize.brute(sumofsquares, ranges, Ns=5, finish=None, full_output=True)

print sumofsquares(fitpars['Brute'])
print "---"
print_values(tuple(fitpars['Brute']))
print "---"
comparefit(tuple(fitpars['Brute']))
```

5.4  Unconstrained Fittings on dAlpha/dT using least squares

```python
def errdiff(p):
    model_alpha = array(integrate_pars(tuple(p)))
    
    model_dalpha = SB_Model(model_alpha, T_C, *p)
    return array((derExperimental_Alpha - model_dalpha)**2)
```

```python
%%time

names = ['A', 'E', 'm', 'n', 'p']
p0 = (300., 40000.0, 0.75, 6.0, 0)

Start_T = 25.0
End_T = 1000.0
dT = 1

Steps = (End_T-Start_T)/dT
Lift_T = 500
Lift_Alpha = 0.37

T_C = (np.linspace(Start_T,End_T,Steps))
T_K = T_C + 273.15

initialalpha = 0.0000000005


Experimental_Alpha = array(gompertz_noise(T_C,End_T,Lift_T,Lift_Alpha, 0))
derExperimental_Alpha = scipy.signal.savgol_filter(Experimental_Alpha, window_length=55, polyorder=2, deriv=2, mode='interp')/dT

fitpars = {}
newstart = numpy.array(p0)
fitpars['Unconstrained dAlpha/dT'], cov_p, infodict, mesg, ierr = scipy.optimize.leastsq(errdiff, newstart, full_output=True)

print sumofsquares(fitpars['Unconstrained dAlpha/dT'])
print "---"
print_values(p0)
print "---"
print_values(fitpars['Unconstrained dAlpha/dT'])
print "---"
comparefit(fitpars['Unconstrained dAlpha/dT'])
```
